```python
X_1 = dataset_1.iloc[:, 0:14]
X_1 = X_1.values
Y_1 = dataset_1.iloc[:, -1].values
from sklearn.model_selection import train_test_split

X_train_1, X_test_1, Y_train_1, Y_test_1 = train_test_split(
    X_1, Y_1, test_size=0.25, stratify=Y_1, random_state=11
)
len(Y_train_1), Y_train_1.sum(), len(Y_test_1), Y_test_1.sum()
```

### Dokonuje sprawdzenia jakoÅ›ci piÄ™ciu klasyfikatorÃ³w takich jak :
    1.KNeighborsClassifier
    2.DecisionTreeClassifier
    3.GradienBoostingClassifer
    4.RandomForestClassifer
    5.GausianNB
    6.LogisticRegresion
    7.MLPClassifer

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier

from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import cross_val_score


models_1 = []
models_1.append(("KNN", KNeighborsClassifier(n_neighbors=3)))
models_1.append(("DSC", DecisionTreeClassifier(random_state=0, max_depth=2)))
models_1.append(
    (
        "GBC",
        GradientBoostingClassifier(
            n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0
        ),
    )
)
models_1.append(("RFC", RandomForestClassifier(max_depth=2, random_state=0)))
models_1.append(("GNB", GaussianNB()))
models_1.append(("LR", LogisticRegression(random_state=0)))
models_1.append(("MLP", MLPClassifier(random_state=1, max_iter=300)))

# evaluate each model in turn
results = []
names = []
for name, model in models_1:
    kfold_1 = RepeatedStratifiedKFold(n_splits=2, n_repeats=2, random_state=0)
    cv_results_1 = cross_val_score(
        model, X_train_1, Y_train_1, cv=kfold_1, scoring="balanced_accuracy"
    )
    results.append(cv_results_1)
    names.append(name)
    msg = "%s: %.2f (%.3f)" % (name, cv_results_1.mean(), cv_results_1.std())
    print(msg)
```
